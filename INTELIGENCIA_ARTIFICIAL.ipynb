```python
from collections import deque
import time

grafo = {
    'A': ['B','C'],
    'B': ['D','E'],
    'C': ['F','G'],
    'D': [],
    'E': [],
    'F': [],
    'G': []
}

def bfs_visual(grafo, inicio):
    visitados = set()
    cola = deque([inicio])

    print("Recorrido BFS:")

    while cola:
        nodo = cola.popleft()

        if nodo not in visitados:
            print("Visitando nodo:", nodo)
            visitados.add(nodo)
            time.sleep(1)

            for vecino in grafo[nodo]:
                cola.append(vecino)

def dfs_visual(grafo, inicio):
    visitados = set()
    pila = [inicio]

    print("\nRecorrido DFS:")

    while pila:
        nodo = pila.pop()

        if nodo not in visitados:
            print("Visitando nodo:", nodo)
            visitados.add(nodo)
            time.sleep(1)

            for vecino in grafo[nodo]:
                pila.append(vecino)


bfs_visual(grafo,'A')
dfs_visual(grafo,'A')
```
