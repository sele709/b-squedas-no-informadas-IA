```python
import networkx as nx
import matplotlib.pyplot as plt
import time
from collections import deque

# Crear el grafo
G = nx.Graph()

edges = [
    ('A','B'),
    ('A','C'),
    ('B','D'),
    ('B','E'),
    ('C','F'),
    ('C','G')
]

G.add_edges_from(edges)

pos = {
    'A': (0,3),
    'B': (-1,2),
    'C': (1,2),
    'D': (-1.5,1),
    'E': (-0.5,1),
    'F': (0.5,1),
    'G': (1.5,1)
}

def dibujar(visitados):
    plt.figure(figsize=(6,4))

    colores = []
    for nodo in G.nodes():
        if nodo in visitados:
            colores.append("red")
        else:
            colores.append("lightblue")

    nx.draw(G, pos,
            with_labels=True,
            node_color=colores,
            node_size=2000,
            font_size=12)

    plt.show()

def bfs(inicio):
    visitados = []
    cola = deque([inicio])

    while cola:
        nodo = cola.popleft()

        if nodo not in visitados:
            visitados.append(nodo)

            print("Visitando:", nodo)
            dibujar(visitados)
            time.sleep(1)

            for vecino in G.neighbors(nodo):
                cola.append(vecino)

def dfs(inicio):
    visitados = []
    pila = [inicio]

    while pila:
        nodo = pila.pop()

        if nodo not in visitados:
            visitados.append(nodo)

            print("Visitando:", nodo)
            dibujar(visitados)
            time.sleep(1)

            for vecino in G.neighbors(nodo):
                pila.append(vecino)

print("RECORRIDO BFS")
bfs('A')

print("RECORRIDO DFS")
dfs('A')
```
